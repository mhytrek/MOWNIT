```python
import scipy.misc
import math
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = []
y_data = []
z_data = []
input = []

for j in range(20):
    s = j/19
    s *= 2*math.pi
    for i in range(20):
        t = i/19
        t *= math.pi

        x = math.cos(s)*math.sin(t)
        x_data.append(x)
        y = math.sin(s)*math.sin(t)
        y_data.append(y)
        z = math.cos(t)
        z_data.append(z)

        input.append([x,y,z])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = np.array(x_data)
xdata = np.array(y_data)
ydata = np.array(z_data)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
plt.show()
```

```python
def make_elipse(input, M):
    elips = []
    for i in input:
        i = np.array(i).reshape(3,1)
        elips.append(M @ i)
```

```python
def get_elipse_from_matrix(matrix, sphere):
    elipse = [
        np.matmul(point, matrix) for point in sphere
    ]
    x_elipse = [point[0] for point in elipse]
    y_elipse = [point[1] for point in elipse]
    z_elipse = [point[2] for point in elipse]

    return x_elipse, y_elipse, z_elipse
```

```python
x_elips, y_elips, z_elips = get_elipse_from_matrix(A1,input)

fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = np.array(x_elips)
xdata = np.array(y_elips)
ydata = np.array(z_elips)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
plt.show()
```

```python
def zad4(matrix, sphere):
    U,S,V = np.linalg.svd(matrix)

    w1 = sphere @ V.transpose()
    w2 = sphere @ (S @ V.transpose())
    w3 = sphere @ (U @ (S @ V.transpose()))

    return w1,w2,w3
```
